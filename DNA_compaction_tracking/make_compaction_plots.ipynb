```python
#dependencies
import pickle
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 20})
from smtools.misc import straight_line
```

## Tracking DNA compaction
----

This notebook will produce average compaction plots for:
* HP1$\alpha$ 50$\mu$M
* HP1$\alpha$ 5$\mu$M
* HP1$\alpha$ 500nM
* HP1$\beta$ 50$\mu$M
* HP1$\gamma$ 50$\mu$M
* HP1$\alpha$-hinge 5$\mu$M
* HP1$\alpha\Delta$CTE 50$\mu$M
* HP1$\alpha\Delta$CTE$\Delta$NTE 50$\mu$M
* HP1$\alpha\Delta$NTE 50$\mu$M


### HP1$\alpha$ 50$\mu$M
----

```python
t,a,s = pickle.load(open("pickled/alpha_50_out.p", "rb")) 

framerate = .1
l,r = 106,115
p, e = curve_fit(straight_line,t[l:r],a[l:r])

fig = plt.figure(figsize = (12,5))
ax1 = fig.add_subplot(111)
ax1.plot(t,straight_line(np.array(t),*p), "c--", linewidth = 5)
ax1.axvline(x=0.1,color = "red", linewidth = 2)
ax1.axvline(x=1,color = "red", linewidth = 2)
ax1.errorbar(t,a, yerr = s, 
             marker = "o", linestyle = "None", 
             errorevery=1, color = "k") 
ax1.set_ylim(-10,50)
ax1.set_xlim(-3.5,5)
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Relative position (kbps)")
ax1.set_title(r'HP1$\alpha$ 50$\mu$M')
plt.show()

print("average velocity: ","{:.2f}".format(p[0]*(48.502/48)),
      "kbp/s")
```
