# Analysis of segmented HP1 phase droplets

This notebook builds out our analysis method for the segmented bright field images of HP1$\alpha$ droplets developed [here](path_to_part_1). This notebook is also provided with 13 test images. 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

from skimage.external.tifffile import imread, imsave
from skimage.feature import canny
from skimage.measure import regionprops,find_contours
from skimage.draw import circle

from scipy import ndimage as ndi
from scipy import fftpack
from scipy.ndimage.filters import convolve

import os
import time
#-- this is here to suppress the warning about regionprops.coords
#-- switching to r,c coords instead of x,y.
import warnings
warnings.filterwarnings("ignore")


def LOG(x,y,sigma):
    a = (-1/(np.pi*sigma**4))
    b = (1-((x**2+y**2)/(2*sigma**2)))
    c = np.exp(-((x**2+y**2)/(2*sigma**2)))
    return(a*b*c)

def mk_LOG_kernel(size, sigma):
    k = np.zeros((size,size))
    
    for i in range(size):
        for j in range(size):
            k[i,j] = LOG(i-(size//2),j-(size//2),sigma)
    scaled = 1./abs(k).min()*k
    return np.rint(scaled)

def mk_bf_segments(Image, method = "summed", sumrange = 10, 
                   kernelsize = 3, kernelsigma = 1.5, Cannysigma = 4,
                   thresh_ecc = 0.97, thresh_area = 50):

    if method == "summed":
        mask = np.zeros_like(Image, dtype=bool)
        for i in range(sumrange):
            edges = canny(Image, sigma=i)
            filled = ndi.binary_fill_holes(edges)
            labels, _ = ndi.label(filled)
            detect = np.zeros_like(Image)
            for i in regionprops(labels):
                if i.eccentricity<thresh_ecc and i.area>thresh_area:
                    for y,x in i.coords:
                        detect[y,x] = 1
            mask = np.logical_or(mask, detect)
        
    if method == "LOG":
        Kernel = mk_LOG_kernel(kernelsize,kernelsigma)
        highpass_filtered = convolve(Image, Kernel)
        edges = canny(highpass_filtered, Cannysigma)
        filled = ndi.binary_fill_holes(edges)
        labels, _ = ndi.label(filled) 
        mask = np.zeros_like(Image)
        for i in regionprops(labels):
            if i.eccentricity<thresh_ecc and i.area>thresh_area:
                for y,x in i.coords:
                    mask[y,x] = 1
    return mask

def bootstrapSurvival(lifetimes, stepsize = 1, 
                      CI = 0.5, samples = 200):
    
    bins = np.linspace(0,max(lifetimes)/stepsize,max(lifetimes)/stepsize)
    n,_ = np.histogram(lifetimes, bins, normed=True)
    measured_cdf = np.cumsum(n)
    measured_cdf[:] = [1.0-x for x in measured_cdf]
    
    tmp_survival = []
    for i in range(samples):
        n,_ = np.histogram(np.random.choice(lifetimes, len(lifetimes)), 
                           bins, normed=True)
        cdf = np.cumsum(n)
        cdf[:] = [1.0-x for x in cdf]
        tmp_survival.append(cdf)
    
    top_err, bot_err = [],[]   
    for i in range(len(bins[:-1])):
        data = [row[i] for row in tmp_survival]
        data.sort()
        top_err.append(data[int(CI*len(data))])
        bot_err.append(data[int((1.-CI)*len(data))])
        
    return bins[:-1],measured_cdf, top_err, bot_err
```

We developed two methods in [the previous notebook]() for segmenting the brightfield images of HP1$\alpha$ droplets, and while there were differences between the two strategies, we wanted to test how each method affected the downstream analysis. For these data, we have decided to use the area of each detected droplet as the statistic to describe the phase separation behavior under varying conditions. The distribution of those areas, 
begin, we read in all of the test images from the provided folder, and segment them using both the summed and LOG methods. We then collect the area of each detected droplet, which is the statistic we have decided to use to describe the phase separation behavior under varying conditions. Below we also print out the runtime of each method per image.

```python
imlist = [imread("test_images/"+i) for i in os.listdir("test_images/")]

LOG_areas, Summed_areas = [], []
for im in range(len(imlist)):
    print("Summed...test_image_", im+1,"   ", end='')
    start = time.time()
    im_seg = mk_bf_segments(imlist[im])
    labels, _ = ndi.label(im_seg) 
    for label in regionprops(labels):
        Summed_areas.append(label.area)
    end = time.time()
    print("runtime = "+"{:.3f}".format(end - start))
    print("LOG......test_image_", im+1,"   ", end='')
    start = time.time()
    im_seg = mk_bf_segments(imlist[im], method="LOG")
    labels, _ = ndi.label(im_seg) 
    for label in regionprops(labels):
        LOG_areas.append(label.area)
    end = time.time()
    print("runtime = "+"{:.3f}".format(end - start))
```

Generally both methods look similar. 

```python
fig = plt.figure(figsize=(16,5))

ax1 = fig.add_subplot(121)
ax1.set_title(r"PDF of Summed areas,  $\langle A \rangle = $"+"{:.2f}".format(np.mean(Summed_areas)), fontsize = 20)
ax1.hist(Summed_areas, 50, log=True, ec = "k", fc = "#008fd5")

ax2 = fig.add_subplot(122)
ax2.set_title(r"PDF of LOG areas,  $\langle A \rangle = $"+"{:.2f}".format(np.mean(LOG_areas)), fontsize = 20)
ax2.hist(LOG_areas, 50, log=True, ec = "k", fc = "#fc4f30")

plt.show()
```

But we will go with the complemntary cumulative distribution function, or the survival probability, because it a more robust statistic. Furthermore we are going to generate confidence intervals from the Bootstrap method. 


![](Combined_plots_small.gif)

```python
import random

Sum_copy = Summed_areas.copy()
for count in range(len(Summed_areas)-len(LOG_areas)):
    Sum_copy.pop(random.choice([i for i,num in enumerate(Sum_copy) if num<500]))

fig = plt.figure(figsize=(18,6))

ax1 = fig.add_subplot(121)

x,y,ery_t,ery_b = bootstrapSurvival(LOG_areas, samples=200)
ax1.semilogy(x,y, linewidth = 5, label = "LOG", color = "#fc4f30")
ax1.fill_between(x,ery_t, ery_b, alpha = .2, color = "#fc4f30")
x,y,ery_t,ery_b = bootstrapSurvival(Summed_areas, samples=200)
ax1.semilogy(x,y, linewidth = 5, label = "Summed", color = "#008fd5")
ax1.fill_between(x,ery_t, ery_b, alpha = .2, color = "#008fd5")

ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.set_xlim(50)
ax1.legend(loc = 1, fontsize = 20)

ax2 = fig.add_subplot(122)

x,y,ery_t,ery_b = bootstrapSurvival(LOG_areas, samples=200)
ax2.semilogy(x,y, linewidth = 5, label = "LOG", color = "#fc4f30")
ax2.fill_between(x,ery_t, ery_b, alpha = .2, color = "#fc4f30")
x,y,ery_t,ery_b = bootstrapSurvival(Summed_areas, samples=200)
ax2.semilogy(x,y, linewidth = 5, label = "Summed", color = "#008fd5")
ax2.fill_between(x,ery_t, ery_b, alpha = .2, color = "#008fd5")
x,y,ery_t,ery_b = bootstrapSurvival(Sum_copy, samples=200)
ax2.semilogy(x,y, linewidth = 5, label = "Censored", color = "#6d904f")
ax2.fill_between(x,ery_t, ery_b, alpha = .2, color = "#6d904f")

ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.set_xlim(50,1750)
ax2.set_ylim(.1,1)
ax2.legend(loc = 1, fontsize = 20)

plt.show()
```

Well it looks like at the larger droplet sizes both methods track together, which is great. However at the smaller sizes, there is a clear differences. The difference isnt due to the lower number of droplets determined by the LOG method, which can be shown by censoring the summed data. Fitting the data.
