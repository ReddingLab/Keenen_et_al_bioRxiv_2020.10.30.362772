```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

from skimage.external.tifffile import imread, imsave
from skimage.feature import canny
from skimage.measure import regionprops,find_contours
from skimage.draw import circle

from scipy import ndimage as ndi
from scipy import fftpack
from scipy.ndimage.filters import convolve

import os
import time
#-- this is here to suppress the warning about regionprops.coords
#-- switching to r,c coords instead of x,y.
import warnings
warnings.filterwarnings("ignore")


def LOG(x,y,sigma):
    a = (-1/(np.pi*sigma**4))
    b = (1-((x**2+y**2)/(2*sigma**2)))
    c = np.exp(-((x**2+y**2)/(2*sigma**2)))
    return(a*b*c)

def mk_LOG_kernel(size, sigma):
    k = np.zeros((size,size))
    
    for i in range(size):
        for j in range(size):
            k[i,j] = LOG(i-(size//2),j-(size//2),sigma)
    scaled = 1./abs(k).min()*k
    return np.rint(scaled)

def mk_bf_segments(Image, method = "summed", sumrange = 10, 
                   kernelsize = 3, kernelsigma = 1.5, Cannysigma = 4,
                   thresh_ecc = 0.97, thresh_area = 50):

    if method == "summed":
        mask = np.zeros_like(Image, dtype=bool)
        for i in range(sumrange):
            edges = canny(Image, sigma=i)
            filled = ndi.binary_fill_holes(edges)
            labels, _ = ndi.label(filled)
            detect = np.zeros_like(Image)
            for i in regionprops(labels):
                if i.eccentricity<thresh_ecc and i.area>thresh_area:
                    for y,x in i.coords:
                        detect[y,x] = 1
            mask = np.logical_or(mask, detect)
        
    if method == "LOG":
        Kernel = mk_LOG_kernel(kernelsize,kernelsigma)
        highpass_filtered = convolve(Image, Kernel)
        edges = canny(highpass_filtered, Cannysigma)
        filled = ndi.binary_fill_holes(edges)
        labels, _ = ndi.label(filled) 
        mask = np.zeros_like(Image)
        for i in regionprops(labels):
            if i.eccentricity<thresh_ecc and i.area>thresh_area:
                for y,x in i.coords:
                    mask[y,x] = 1
    return mask

def bootstrapSurvival(lifetimes, stepsize = 1, 
                      CI = 0.5, samples = 200):
    
    bins = np.linspace(0,max(lifetimes)/stepsize,max(lifetimes)/stepsize)
    n,_ = np.histogram(lifetimes, bins, normed=True)
    measured_cdf = np.cumsum(n)
    measured_cdf[:] = [1.0-x for x in measured_cdf]
    
    tmp_survival = []
    for i in range(samples):
        n,_ = np.histogram(np.random.choice(lifetimes, len(lifetimes)), 
                           bins, normed=True)
        cdf = np.cumsum(n)
        cdf[:] = [1.0-x for x in cdf]
        tmp_survival.append(cdf)
    
    top_err, bot_err = [],[]   
    for i in range(len(bins[:-1])):
        data = [row[i] for row in tmp_survival]
        data.sort()
        top_err.append(data[int(CI*len(data))])
        bot_err.append(data[int((1.-CI)*len(data))])
        
    return bins[:-1],measured_cdf, top_err, bot_err
```

```python
imlist = [imread("test_images/"+i) for i in os.listdir("test_images/")]

LOG_areas, Summed_areas = [], []
for im in range(len(imlist)):
    print("Summed...test_image_", im+1,"   ", end='')
    start = time.time()
    im_seg = mk_bf_segments(imlist[im])
    labels, _ = ndi.label(im_seg) 
    for label in regionprops(labels):
        Summed_areas.append(label.area)
    end = time.time()
    print("runtime = "+"{:.3f}".format(end - start))
    print("LOG......test_image_", im+1,"   ", end='')
    start = time.time()
    im_seg = mk_bf_segments(imlist[im], method="LOG")
    labels, _ = ndi.label(im_seg) 
    for label in regionprops(labels):
        LOG_areas.append(label.area)
    end = time.time()
    print("runtime = "+"{:.3f}".format(end - start))
```

```python
from scipy.optimize import curve_fit

def stretched_exponential(x, k0, A, B):
    func = np.exp(- k0*(x**(B)))
    return func

def double_exponential(x, a0, k0, a1, k1):
    func = a0 * np.exp(- k0*x) + \
           a1 * np.exp(- k1*x)
    return func

fig = plt.figure(figsize=(16,6))

ax1 = fig.add_subplot(121)


x,y,ery_t,ery_b = bootstrapSurvival(Summed_areas, samples=200)
ax1.semilogy(x,y, linewidth = 5, label = "Summed", color = "#e5ae38")
ax1.fill_between(x,ery_t, ery_b, alpha = .2, color = "#e5ae38")

truncx, truncy = [],[]
for i,j in zip(x,y):
    if j>0.01:
        truncx.append(i)
        truncy.append(j)
    else:
        break
        
popt, pcov = curve_fit(double_exponential, truncx, truncy)
ax1.plot(x, double_exponential(x,*popt), label = "exponential2/3")
print(popt)
    
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.set_xlim(50,6000)
ax1.set_ylim(.01,1)
ax1.legend(loc = 1, fontsize = 20)

plt.show()
```

```python
from scipy.optimize import curve_fit

def exponential(x, a0, k0):
    func = a0* np.exp(- k0*x)
    return func

def double_exponential(x, a0, k0, a1, k1):
    func = a0 * np.exp(- k0*x) + \
           a1 * np.exp(- k1*x)
    return func

def triple_exponential(x, a0, k0, a1, k1, a2, k2):
    func = a0 * np.exp(- k0*x) + \
           a1 * np.exp(- k1*x) + \
           a2 * np.exp(- k2*x)
    return func

def tetra_exponential(x, a0, k0, a1, k1, 
                      a2, k2, a3, k3):
    func = a0 * np.exp(- k0*x) + \
           a1 * np.exp(- k1*x) + \
           a2 * np.exp(- k2*x) + \
           a3 * np.exp(- k3*x)
    return func

def penta_exponential(x, a0, k0, a1, k1, 
                      a2, k2, a3, k3, 
                      a4, k4):
    func = a0 * np.exp(- k0*x) + \
           a1 * np.exp(- k1*x) + \
           a2 * np.exp(- k2*x) + \
           a3 * np.exp(- k3*x) + \
           a4 * np.exp(- k4*x)
    return func


fig = plt.figure(figsize=(16,6))

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)


x,y,ery_t,ery_b = bootstrapSurvival(Summed_areas, samples=200)
ax1.semilogy(x,y, linewidth = 5, label = "Summed", color = "#e5ae38")
ax1.fill_between(x,ery_t, ery_b, alpha = .2, color = "#e5ae38")

for i, labels in zip([exponential, double_exponential, 
                      triple_exponential, tetra_exponential], 
                     ["exponential", "double_exponential", 
                      "triple_exponential", "tetra_exponential"]):
    popt, pcov = curve_fit(i, x, y)
    ax1.plot(x, i(x,*popt), label = labels)
    
    pos = [i for i in range(len(popt[1::2]))]
    plt.plot(pos,popt[1::2], marker = "o", markersize = 10)

    
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.set_xlim(50,6000)
ax1.set_ylim(.01,1)
#ax2.set_xlim(-.5,1.5)
#ax2.set_ylim(-.0025,0.01)
ax1.legend(loc = 1, fontsize = 20)

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

class IndexTracker(object):
    def __init__(self, ax, X):
        self.ax = ax
        ax.set_title('use scroll wheel to navigate images')

        self.X = X
        rows, cols, self.slices = X.shape
        self.ind = 0

        self.im = ax.imshow(self.X[:, :, self.ind])
        self.update()

    def onpress(self, event):
        #print("%s %s" % (event.button, event.step))
        if event.key == 'right':
            self.ind = (self.ind + 1) % self.slices
        elif event.key == 'left':
            self.ind = (self.ind - 1) % self.slices
        
        self.update()

        
    def onscroll(self, event):
        print("%s %s" % (event.button, event.step))
        if event.button == 'up':
            self.ind = (self.ind + 1) % self.slices
        else:
            self.ind = (self.ind - 1) % self.slices
        self.update()
        
        
    def update(self):
        self.im.set_data(self.X[:, :, self.ind])
        ax.set_ylabel('slice %s' % self.ind)
        self.im.axes.figure.canvas.draw()


fig, ax = plt.subplots(1, 1)

X = np.random.rand(20, 20, 40)
X1 = np.stack(imlist, axis=2)
print(X.shape,X1.shape)


tracker = IndexTracker(ax, X1)


fig.canvas.mpl_connect('key_press_event', tracker.onpress)
fig.canvas.mpl_connect('scroll_event', tracker.onscroll)
plt.show()
```
