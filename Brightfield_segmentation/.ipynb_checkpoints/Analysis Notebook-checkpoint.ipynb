# Analysis Notebook


In this notebook we succesively open and analyze the brightfield droplet data collected with HP1$\alpha$ under various conditions. Each section follows the following structure:
* Open images from a condition or range of conditions
* detect and measure area and perimeter of droplets
* write out to a data file

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

from skimage.external.tifffile import imread, imsave
from skimage.feature import canny
from skimage.measure import regionprops,find_contours
from skimage.draw import circle
from skimage.color import rgb2gray

from scipy import ndimage as ndi
from scipy import fftpack
from scipy.ndimage.filters import convolve

import re
import pandas as pd

import os
import time
#-- this is here to suppress the warning about regionprops.coords
#-- switching to r,c coords instead of x,y.
import warnings
warnings.filterwarnings("ignore")


def LOG(x,y,sigma):
    a = (-1/(np.pi*sigma**4))
    b = (1-((x**2+y**2)/(2*sigma**2)))
    c = np.exp(-((x**2+y**2)/(2*sigma**2)))
    return(a*b*c)

def mk_LOG_kernel(size, sigma):
    k = np.zeros((size,size))
    
    for i in range(size):
        for j in range(size):
            k[i,j] = LOG(i-(size//2),j-(size//2),sigma)
    scaled = 1./abs(k).min()*k
    return np.rint(scaled)

def mk_bf_segments(Image, fft = True, fft_thresh = 150, sumrange = 10, 
                   kernelsize = 3, kernelsigma = 1.5, Cannysigma = 4,
                   thresh_ecc = 0.97, thresh_area = 50):
    

    if fft == True:
        im_fft = fftpack.fft2(Image)
        im_fft2 = im_fft.copy()
        row, col = im_fft2.shape
        
        for h in range(row):
            val1 = np.sqrt(fft_thresh**2 - h**2)
            val2 = np.sqrt(fft_thresh**2 - (h-row)**2)

            if np.isnan(val1) and np.isnan(val2):
                im_fft2[h,:] = 0    
            elif np.isnan(val2) and not np.isnan(val1):
                im_fft2[h,int(val1):col-int(val1)] = 0
            elif np.isnan(val1) and not np.isnan(val2):
                im_fft2[h,int(val2):col-int(val2)] = 0
        
        Image = fftpack.ifft2(im_fft2).real

    mask = np.zeros_like(Image, dtype=bool)
    for i in range(sumrange):
        edges = canny(Image, sigma=i)
        filled = ndi.binary_fill_holes(edges)
        labels, _ = ndi.label(filled)
        detect = np.zeros_like(Image)
        for i in regionprops(labels):
            if i.eccentricity<thresh_ecc and i.area>thresh_area:
                for y,x in i.coords:
                    detect[y,x] = 1
        mask = np.logical_or(mask, detect)
        
    return mask

def bootstrapSurvival(lifetimes, stepsize = 1, 
                      CI = 0.5, samples = 200):
    
    bins = np.linspace(0,max(lifetimes)/stepsize,max(lifetimes)/stepsize)
    n,_ = np.histogram(lifetimes, bins, normed=True)
    measured_cdf = np.cumsum(n)
    measured_cdf[:] = [1.0-x for x in measured_cdf]
    
    tmp_survival = []
    for i in range(samples):
        n,_ = np.histogram(np.random.choice(lifetimes, len(lifetimes)), 
                           bins, normed=True)
        cdf = np.cumsum(n)
        cdf[:] = [1.0-x for x in cdf]
        tmp_survival.append(cdf)
    
    top_err, bot_err = [],[]   
    for i in range(len(bins[:-1])):
        data = [row[i] for row in tmp_survival]
        data.sort()
        top_err.append(data[int(CI*len(data))])
        bot_err.append(data[int((1.-CI)*len(data))])
        
    return bins[:-1],measured_cdf, top_err, bot_err
```

```python
Path = "/Volumes/SyAnylzssss/2019Phasedata/HP1a/601DNA/"

DNA_dirs = ["400ngul601DNA","200ngul601DNA","100ngul601DNA","50ngul601DNA",
            "25ngul601DNA","12.5ngul601DNA","6.25ngul601DNA","3.125ngul601DNA",
            "1.56ngulDNA","0.750ngul601DNA","0.360ngul601DNA","0ngul601DNA"]

DNAconcentrations = ["400","200","100","50",
                     "25","12.5","6.25","3.125",
                     "1.56","0.750","0.360","0"]

DNAid = "601 DNA"
```

```python

for DNA_dir,DNA_conc in zip(DNA_dirs,DNAconcentrations):

    print(DNA_conc)
    print("---------------------")
    new_path = Path+DNA_dir
    for concentration in os.listdir(new_path):
        print(concentration)
        for i in os.listdir(new_path+"/"+concentration):

            areas = []
            perimeters = []
            eccentricities = []
            extents = []

            for image in os.listdir(new_path+"/"+concentration+"/"+i):
                print("working on ...   ", image,end='\r')
                im = imread(new_path+"/"+concentration+"/"+i+"/"+image)
                im = rgb2gray(im)

                im_seg = mk_bf_segments(im)
                labels, _ = ndi.label(im_seg) 
                for label in regionprops(labels):

                    areas.append(label.area)
                    eccentricities.append(label.eccentricity)
                    perimeters.append(label.perimeter)
                    extents.append(label.extent)

                fig = plt.figure(figsize=(6,6))
                plt.imshow(im, cmap=plt.cm.gray)
                contours = find_contours(im_seg, .1)
                for n, contour in enumerate(contours):
                    plt.plot(contour[:, 1], contour[:, 0], linewidth=2)
                plt.axis('off')
                plt.savefig("results/segmented_images/"+DNA_conc+"601DNA__"+concentration+"__"+image)

            conc = re.findall(r"[-+]?\d*\.\d+|\d+",concentration)[0]
            HP1concentration = [conc for index in areas] 
            DNAconcentration = [DNA_conc for index in areas]
            filename = [image for index in areas]
            DNAid = ["601 DNA" for index in areas]
            data = dict([('area', areas), ('perimeter', perimeters), ('eccentricity', eccentricities), 
                         ('extent', extents), ("HP1 concentration", HP1concentration), 
                         ("DNA concentration", DNAconcentration),("DNA ID", DNAid),
                         ("filename",filename)])   

            try:
                data_out = data_out.append(pd.DataFrame(data))
            except NameError:
                data_out = pd.DataFrame(data)
            
            
        data_out.to_csv("results/results_"+DNA_conc+"_"+concentration+".dat", sep='\t')

        print()
```
